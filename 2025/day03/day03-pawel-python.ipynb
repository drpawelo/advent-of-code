```python
def load_file_lines(filename):
    lines = []
    file = open(filename)
    for line in file:
        lines.append(line.rstrip("\n"))
    file.close()
    return lines

datas = load_file_lines("./pawel.txt")
# datas
tests = load_file_lines("./test.txt")
tests
```

```python


def highest_two_numbers(some_number):
    digits = [int(letter) for letter in str(some_number)]
    largest_not_last = max(digits[:-1])
    index_of_largest_not_last = digits.index(largest_not_last)
    largest_after = max(digits[index_of_largest_not_last+1:])
    index_of_largest_after = digits.index(largest_after)
    return int(f"{digits[index_of_largest_not_last]}{digits[index_of_largest_after]}")
print(highest_two_numbers("987654321111111"))
print(highest_two_numbers(811111111111119))
print(highest_two_numbers(234234234234278))
print(highest_two_numbers(818181911112111))
```

```python
def run_file(file_datas):
    answers_all = []
    for data in file_datas:
        answers_all.append(highest_two_numbers(data))
    print(answers_all)
    return sum(answers_all)

run_file(tests)
```

```python


def highest_n_numbers(some_number, needed_digits=12):
    # print(some_number)
    digits = [int(letter) for letter in str(some_number)]
    result_digits = []
    index_of_min_allowed = 0
    while len(result_digits) < needed_digits:
        # leave space for the tail and find first allowed largest. 
        # eg. if we have 10 digits, need 5, and this is the 1st one. it has to be betwen index 0 and 5
        # eg. if we have 10 digits, need 5, and this is the 2nd one. (first was at index 2) it has to be betwen index 3 and 6
        # so min is index_of_prev_picked +1. and max is: len(digits) - how many numbers I still need
        max_allowed_index = len(digits) - (needed_digits - len(result_digits)) +1
        # print('search in', digits[index_of_min_allowed:max_allowed_index])
        largest_allowed = max(digits[index_of_min_allowed:max_allowed_index])
        index_of_largest_allowed = digits[index_of_min_allowed:max_allowed_index].index(largest_allowed) + index_of_min_allowed
        result_digits.insert(len(result_digits), str(digits[index_of_largest_allowed]))
        # print("s e w i res")
        # print(index_of_min_allowed, max_allowed_index, largest_allowed, index_of_largest_allowed,result_digits)
        # inrease left needle by one, or to where recent winner was
        index_of_min_allowed = max(index_of_largest_allowed+1,index_of_min_allowed+1)
    return int("".join(result_digits))

# print(highest_n_numbers("987654321111111"))
# print(highest_n_numbers(811111111111119))
print(highest_n_numbers(234234234234278)) 
print(434234234278, "exp")
# 234234234234278
# ..4.34234234278
#   4 34234234278
# print(highest_n_numbers(818181911112111))
```

```python
def run_file(file_datas):
    answers_all = []
    for data in file_datas:
        answers_all.append(highest_n_numbers(data))
    print(answers_all)
    return sum(answers_all)

run_file(tests)
```
