```python
def load_file_lines(filename):
    lines = []
    file = open(filename)
    for line in file:
        lines.append(line.rstrip("\n"))
    file.close()
    return lines

datas = load_file_lines("./pawel.txt")
# datas
tests = load_file_lines("./test.txt")
tests
```

```python

class Dial:
    def __init__(self):
        self.value = 50
        self.number_of_zeros = 0

    def move(self, direction, distance, loud=False):
        self.value += distance if direction == 'R' else -distance
        self.value = self.value % 100
        self.number_of_zeros += 1 if self.value == 0 else 0
        if loud:
            print(self)

    def parse_direction(move_string):
        direction = move_string[0]
        distance = int(move_string[1:])
        return (direction, distance)

    def __repr__(self):
        return f'valie: {self.value:2} - score: {self.number_of_zeros}'
```

```python
# test parsing
Dial.parse_direction("R12")
Dial.parse_direction("L77")
```

```python
# test movements
dial = Dial()
print(dial)
dial.move('R', 10, loud=True)
dial.move('R', 40, loud=True)
```

```python
# test movements with parsing
dial = Dial()
print(dial)
dial.move( *Dial.parse_direction("R10"), loud=True)
dial.move( *Dial.parse_direction("R40"), loud=True)
```

```python
# pass the tests
dial = Dial()
for test in tests:
    dial.move( *Dial.parse_direction(test), loud=True)

assert dial.number_of_zeros == 3
print('test passed')
```

```python
# pass the tests
dial = Dial()
for data in datas:
    dial.move( *Dial.parse_direction(data))

print(dial)
```

```python
class Dial2:
    def __init__(self, start_at = 50):
        self.value = start_at
        self.number_of_zeros = 0

    def move(self, direction, distance, loud=False):
        new_value = self.value + distance * (1 if direction == 'R' else - 1)
        
        if (new_value >= 100 ):
            self.number_of_zeros += new_value // 100
        elif (new_value <= 0 ):
            self.number_of_zeros += (100 - new_value) // 100  # -30 - same as 130
            if (self.value == 0 ): # if started at 0, do not count that 0
                self.number_of_zeros -= 1 

 
        if loud:
            print(f"{direction}{distance}: {self.value:2} -> {new_value:2} - score: {self.number_of_zeros}")

        self.value = new_value % 100


    def parse_direction(move_string):
        direction = move_string[0]
        distance = int(move_string[1:])
        return (direction, distance)

    def __repr__(self):
        return f'value: {self.value:2} - score: {self.number_of_zeros}'

dial = Dial2()
print(dial)
dial.move( *Dial2.parse_direction("R55"), loud=True)
dial.move( *Dial2.parse_direction("R260"), loud=True)
# dial.move( *Dial2.parse_direction("R5"), loud=True) 
# dial.move( *Dial2.parse_direction("L15"), loud=True) #tick
# dial.move( *Dial2.parse_direction("R5"), loud=True) #tick
# dial.move( *Dial2.parse_direction("R5"), loud=True) 
```
