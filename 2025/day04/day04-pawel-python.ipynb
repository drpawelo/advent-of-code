```python
import time
from IPython.display import clear_output


class Map:
    def __init__(self, file_contents):
        self.grid = file_contents
        self.size = len(file_contents)

    def is_full(self, row, column):
        # @ means yes, . means no
        if(row >= 0 and row < len(self.grid) and column >= 0 and column < len(self.grid[0]) ):
            return self.grid[row][column] == "@"
        else:
            return False
        
    

    def neighbours_of(self, row, column):
        neighbours = [
            (a_row, a_column)
            for a_row in range(row-1, row+2)
            for a_column in range(column-1, column+2)
            if not (a_row == row and a_column == column)
        ]
        return neighbours

    def score_of_neighbours_of(self, row, column):
        neighbour_values = [
            1
            for neightbour in self.neighbours_of(row, column)
            if self.is_full(neightbour[0], neightbour[1])
        ]
        return sum(neighbour_values)
    
    def locations_of_items_with_less_then(self, less_than=4):
        items_which_meet_criteria  = [
            (a_row,a_column)
            for a_row in range(len(self.grid))
            for a_column in range(len(self.grid[0]))
            if self.score_of_neighbours_of(a_row,a_column) < less_than and self.is_full(a_row,a_column)
           ]
        # print(items_which_meet_criteria)
        return items_which_meet_criteria
    
    def numbers_of_items_with_less_then(self, less_than=4):
        return len(self.locations_of_items_with_less_then(less_than))

    def replace_string(self, in_string, with_string, at_index):
        return f"{in_string[:at_index]}{with_string}{in_string[at_index+1:]}"

    def empty_location(self, location):
        self.grid[location[0]] = self.replace_string(self.grid[location[0]], ".",location[1] )

    def empty_at_locations(self, locations_to_empty):
        for location in locations_to_empty:
            self.empty_location(location)
    
    def run_generations(self):
        all_emptied_locations = []
        recently_emptied = 1
        while recently_emptied > 0:
            clear_output(wait=True)
            print("\n".join(self.grid))
            time.sleep(0.1)
            items_to_empty = map.locations_of_items_with_less_then()
            recently_emptied = len(items_to_empty)
            all_emptied_locations.extend(items_to_empty)
            self.empty_at_locations(items_to_empty)
            # print("emptied", recently_emptied)
        return len(all_emptied_locations)




map = Map([] + tests)
map.run_generations()
# row = 0
# column = 0
# print(map.neighbours_of(row, column))
# print(map.is_full(row, column))
# print(map.score_of_neighbours_of(row, column))
# # print(map.score_of_neighbours_of(0,2))
# print(map.locations_of_items_with_less_then())
# print(map.numbers_of_items_with_less_then())
```
