```python
def load_file_lines(filename):
    lines = []
    file = open(filename)
    for line in file:
        lines.append(line.rstrip("\n"))
    file.close()
    return lines

datas = load_file_lines("./pawel.txt")
# datas = load_file_lines("./test.txt")
# datas
```

```python
import json

def parse_lines_into_pairs(lines):
    pairs = []
    pair_now = []
    for line in lines:
#         print(f"line {line}")
        if line == '':
#             print(f"skip")
            continue
        else:
#             print(f"add")
            pair_now.append(json.loads(line))
        if len(pair_now) >= 2:
            pairs.append(pair_now)
            pair_now = []
#         print(f"pairs {pairs}")
    return pairs

pairs = parse_lines_into_pairs(datas)
pairs
```

```python
# left needs to be smaller
# left needs to be smaller
# 
class Left:
    smaller = -1
    same = 0
    larger = 1
    def __init__(self):
        pass
    
    def to_bool(number):
        return number == Left.smaller or number == Left.same

def complare_numbers(num1, num2):
    if  num1 < num2:
        return Left.smaller
    elif  num1 == num2:
        return Left.same
    else:
        return Left.larger
    
def is_left_smaller(item1, item2):
#     print(f"items 1: {item1},{item2}")
    if type(item1) is int and  type(item2) is int:
        return complare_numbers(item1, item2)
    
    if type(item1) is int:
        item1 = [item1]
    if type(item2) is int:
        item2 = [item2]
    
    if (len(item2) == 0) and len(item1) > len(item2):
        return Left.larger
    if (len(item1) == 0) and len(item1) < len(item2):
        return Left.smaller
    
#     print(f"items 2: {item1},{item2}")
    if len(item1) != 0 and len(item2) != 0:
        first1 = item1.pop(0)
        first2 = item2.pop(0)
        first_in_order = is_left_smaller(first1, first2)
        if first_in_order == Left.larger or first_in_order == Left.smaller:
            return first_in_order
        else:
            return is_left_smaller(item1, item2)
    else:
        return Left.same

        
# print(is_left_smaller([],[]))
# print(is_left_smaller([1,2,3],[2,3,4]))
# print(is_left_smaller([1,2,3],[2,3,1]))
# print(is_left_smaller([1,8],[4,7,6]))
print(is_left_smaller(6,[4,7,6]))
```

```python
import json

def parse_lines_into_packets(lines):
    packets = []
    for line in lines:
        if line == '':
            continue
        else:
            packets.append(json.loads(line))

    return packets

packets = parse_lines_into_packets(datas)
packets
```

```python
def compare_pairs(source1, source2):
    item1, item2 = json.loads( json.dumps(source1)), json.loads( json.dumps(source2))
    result = is_left_smaller(item1, item2)
    return result

import functools
packets = parse_lines_into_packets(datas)
start = [[2]]
end = [[6]]
packets.extend([start, end])
sorted_packets = sorted(packets, key=functools.cmp_to_key(compare_pairs))

index_of_start = sorted_packets.index(start) +1
index_of_end = sorted_packets.index(end) +1
result = index_of_start * index_of_end
print(f"{result} = {index_of_start} {index_of_end}")

# part 2 was 20383
```
