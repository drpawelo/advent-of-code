```python
from IPython.display import clear_output
import time
```

```python
def load_file_lines(filename):
    lines = []
    file = open(filename)
    for line in file:
        lines.append(line.rstrip("\n"))
    file.close()
    return lines

datas = load_file_lines("./pawel.txt")
# datas = load_file_lines("./test.txt")
# datas
```

```python
rope_part_locations = []
tail_tracks = []

directions_xy = {
    'R':[1,0],
    'L':[-1,0],
    'U':[0,-1],
    'D':[0,1],
}

grid = []

def empty_grid(size):
    grid = []
    for index_x in range(size):
        grid.append([])
        for index_y in range(size):
            grid[-1].append("░")
    return grid

def print_grid(grid_size, tail_tracks, rope_part_locations):
    grid_now = empty_grid(grid_size)
    grid_string = ""

#     grid_now[len(grid_now)//2][len(grid_now)//2] = 's'
        
    for track in tail_tracks:
        if track[0] < len(grid_now) and track[1] < len(grid_now):
            grid_now[track[0]][track[1]] = '▒'
        
    for index in range(len(rope_part_locations)-1,-1,-1):
        row, column = rope_part_locations[index]
        if index == 0:
            grid_now[row][column] = "@"
        else:
            grid_now[row][column] = str(index) 
            
    box_row, box_column = rope_part_locations[-1]
    grid_now[box_row][box_column] = "█"

    
    for row in range(len(grid_now)):
        for column in range(len(grid_now)):
            
            tile = grid_now[column][row]
 
            grid_string += tile
        grid_string += "\n"
    print(grid_string)
    
# grid = empty_grid(30)
# print_grid(len(grid), tail_tracks, rope_part_locations)
```

```python
# rope_part_locations, grid, tail_tracks = start_positions(100)

def start_positions(size, number_of_line_segments):
    middle = size//2
    rope_part_locations = [[middle,middle] for knot in range(number_of_line_segments)]

    return (rope_part_locations, empty_grid(size), [[middle,middle]] )

# [2,2] [-1,0] -> [1,2]
def add_coords(coord1, coord2):
    return [coord1[0]+coord2[0],coord1[1]+coord2[1]]

def distance_of(coord1, coord2):
    return [coord1[0]-coord2[0],coord1[1]-coord2[1]]


# E123456789
def move_all_parts(direction, distance, rope_part_locations):
    for step in range(distance):
        rope_part_locations[0] = add_coords(rope_part_locations[0], directions_xy[direction])
        for segment_index in range(1,len(rope_part_locations)): # 1
            pulling = rope_part_locations[segment_index-1]
            pulled = tail_follows(pulling, rope_part_locations[segment_index])
            rope_part_locations[segment_index] = pulled
        add_to_tracks(rope_part_locations[-1])
        time.sleep(0.05)
        clear_output(wait=True)
        print_grid(len(grid), tail_tracks, rope_part_locations)

    return rope_part_locations
    
def tail_follows(head_position, tail_position):
    distance_head_to_tail = distance_of(head_position, tail_position)
    tail_position = add_coords(tail_position, movement_of_tail(distance_head_to_tail))
    return tail_position

def movement_of_tail(distance):
    if abs(distance[0])< 2 and abs(distance[1])< 2 : # don't move
        return [0,0]
    if abs(distance[0]) >= 2 and abs(distance[1]) < 2:
        move_x = 1 if distance[0] > 0 else -1
        return [move_x ,distance[1]]
    if abs(distance[1]) >= 2 and abs(distance[0]) < 2:
        return [distance[0], 1 if distance[1] > 0 else -1 ]
    if abs(distance[1]) >= 2 and abs(distance[0]) >= 2:
        return [1 if distance[0] > 0 else -1, 1 if distance[1] > 0 else -1 ]
```

```python
# tail_follows([2,1], [0,0])    
def add_to_tracks(tail):
    tail_tracks.append(tail)
    
def as_command(line):# "L 5" --> ['L',5]
    coords = line.split(" ")
    return [coords[0], int(coords[1])]

def number_of_unique_tiles(tiles_coords):
    as_strings = [
    f"{tile[0]}-{tile[1]}"
    for tile in tiles_coords]
    as_strings = list(set(as_strings))
#     print(as_strings)
    return len(as_strings)


# grid = empty_grid(30)

rope_part_locations, grid, tail_tracks = start_positions(70, 10)

print_grid(len(grid), tail_tracks, rope_part_locations)

# rope_part_locations, grid, tail_tracks = start_positions(500, 10) 
# to actually sove the puzzle you need larger grid, like above

for line in datas[0:300]:
    direction, distance = as_command(line)
    rope_part_locations = move_all_parts(direction, distance, rope_part_locations)
    
# print_grid(grid, tail_tracks, rope_part_locations)

print(number_of_unique_tiles(tail_tracks))
```
