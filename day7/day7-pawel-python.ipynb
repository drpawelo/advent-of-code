```python
current_directory_components = [] # glogal variable ["\","downloads","january"] for '\downloads\january'
all_objects = []

def folder_sizes(filename):
    datas = load_file_lines(f"./{filename}")
    map_file_system(datas)
    sizes = sizes_of_folders_smaller_than(all_objects, 1000000000000)
    return sizes

all_sizes = folder_sizes("pawel.txt")
# whole_analysis("example.txt")

def smalles_satisfying_deletion(sizes):
    top_folder_size = max(sizes)
    possible_deletions = []
    for size in sizes:
    #     print(f"{size} \t {top_folder_size - size} \t{top_folder_size - size < total_space - needed_space}")
        if top_folder_size - size < total_space - needed_space:
            possible_deletions.append(size)
    return min(possible_deletions)

smalles_satisfying_deletion(all_sizes)
```
