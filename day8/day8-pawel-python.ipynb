```python
def load_file_lines(filename):
    lines = []
    file = open(filename)
    for line in file:
        lines.append(line.rstrip("\n"))
    file.close()
    return lines

datas = load_file_lines("./pawel.txt")
datas = load_file_lines("./test.txt")
datas
```

```python
def make_grid(file_lines):
    grid = [
        []
        for line in file_lines
    ]
    
    for line_index in range(len(file_lines)):
        for column_index in range(len(file_lines[line_index])):
            grid[line_index].append( int(file_lines[line_index][column_index]))

    return grid

# datas = load_file_lines("./test.txt")
grid = make_grid(datas)
```

```python
def show_pretty(grid):
    result = ''
    for row in grid:
        for item in row:
            result += str(item)
        result += "\n"
    print(result)

show_pretty(grid)
```

```python
def is_visible_1d(trees, index):
    candidate_tree = trees[index]
    visible_below = True
    visible_above = True
    for from_edge in range(0,index) :
        if trees[from_edge] >= candidate_tree:
            visible_below = False
            
    for from_edge in range(index+1,len(trees)) :
        if trees[from_edge] >= candidate_tree:
            visible_above = False
            
#     print(visible_below, visible_above)
    return visible_below or visible_above

print(is_visible_1d([7, 5, 3, 3, 2], 3))
```

```python
def get_row(grid, index):
    return grid[index]
    
def get_column(grid, index):
    return [
        row[index]
        for row in grid
    ]

show_pretty(grid)
print(get_row(grid, 3))
print(get_column(grid, 3))
```

```python
def is_tree_visible(grid, which_row, which_column):
    row = get_row(grid, which_row)
    visible_in_row = is_visible_1d(row,which_column)
    column = get_column(grid, which_column)
    visible_in_column = is_visible_1d(column,which_row)
    return visible_in_row or visible_in_column

show_pretty(grid)
is_tree_visible(grid, 2,2)
```

```python
def visibility_of_trees(grid):
    visibilities = [
        []
        for line in grid
    ]
    
    for line_index in range(len(grid)):
        for column_index in range(len(grid[line_index])):
            is_visible = is_tree_visible(grid, line_index, column_index)
            visibilities[line_index].append( 1 if is_visible else 0 )

    return visibilities

visibilities = visibility_of_trees(grid)
show_pretty(grid)
show_pretty(visibilities)
```

```python
def count_visibles(grid):
    visibilities = visibility_of_trees(grid)
    return sum([
        sum(vis_row)
        for vis_row in visibilities
    ])

count_visibles(grid)
```

```python
# part 2

datas = load_file_lines("./pawel.txt")
# datas = load_file_lines("./test.txt")
grid = make_grid(datas)
```

```python
def scenic_score_1d(trees, index):
    candidate_tree = trees[index]
    visible_below = 0
    visible_above = 0
    for from_edge in range(index-1,-1,-1) :
        visible_below += 1
        if trees[from_edge] >= candidate_tree:
            break
            
    
    for from_edge in range(index+1,len(trees)) :
        visible_above += 1
        if trees[from_edge] >= candidate_tree:
            break
            
#     print(visible_below, visible_above)
    return visible_below * visible_above

print(scenic_score_1d([3, 3, 5, 4, 9], 2))
```

```python
def scenic_score_of_tree(grid, which_row, which_column):
    row = get_row(grid, which_row)
    score_in_row = scenic_score_1d(row,which_column)
    column = get_column(grid, which_column)
    score_in_column = scenic_score_1d(column,which_row)
#     print("scores:",score_in_row, score_in_column)
    return score_in_row * score_in_column

show_pretty(grid)
scenic_score_of_tree(grid, 3,2)
```

```python
def scores_of_all_trees(grid):
    scores = [
        []
        for line in grid
    ]
    
    for line_index in range(len(grid)):
        for column_index in range(len(grid[line_index])):
            score = scenic_score_of_tree(grid, line_index, column_index)
            scores[line_index].append( score )

    return scores

scores = scores_of_all_trees(grid)
show_pretty(grid)
show_pretty(scores)
```

```python
def max_score(grid):
    scores = scores_of_all_trees(grid)
    return max([
        max(row)
        for row in scores
    ])

max_score(grid)
```
